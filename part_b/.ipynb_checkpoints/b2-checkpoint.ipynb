```python
torch.manual_seed(123)

# Define the RBF kernel
def RBF_kernel(x, x_prime, varaince, lengthscale):
    dists_sq = torch.cdist(x, x_prime, p=2).pow(2)
    return (
        varaince*torch.exp(-0.5*dists_sq/(lengthscale**2))
    )

i_s = torch.linspace(1,101,101)
ell_x = torch.tensor(101)

# define a function for computing x
def xi(x):
    return(
         (x-1)/(ell_x-1)
    )
x = xi(i_s).reshape(-1,1)

# define function g
def g(x):
    return(
        -(torch.sin(6*torch.pi*x))**2 + 6*x**2-5*x**4 + (3/2)
    )

# Parameters
ell_kernel = 0.35740169882774353  # Lengthscale from MAP
sigma_sq = 1.912818193435669  # Variance from MAP

# define the weights w
W = torch.zeros(ell_x)
W[0], W[-1] = 1/(2*ell_x-2), 1/(2*ell_x-2)
W[1:-1] = 1/(ell_x-1)
W = W.reshape(-1,1) # ensure W is a column vector 

# compute the kernel matrix induced by x
K = RBF_kernel(x, x, sigma_sq, ell_kernel)

# compute the conditional covariance matrix 
KW = K @ W
WKW = W.T @ K @ W
sigma_f = K - KW @ torch.inverse(WKW) @ KW.T

# Store the different values of q_hat we are given
q_hat = torch.Tensor([0,5,10])

# compute the conditional mean for each of the q_hat values
mu_f_q1 = KW @ torch.inverse(WKW)*q_hat[0]
mu_f_q2 = KW @ torch.inverse(WKW)*q_hat[1]
mu_f_q3 = KW @ torch.inverse(WKW)*q_hat[2]
# Ensure the correct dimensions of the three mean vectors
mu_f_q1 = mu_f_q1.reshape(-1, )
mu_f_q2 = mu_f_q2.reshape(-1, )
mu_f_q3 = mu_f_q3.reshape(-1, ) 
```

```python
# check if sigma_f is symmetric, which is has to be in order to be a valid covariance matrix. 
# But it might not be, due to numerical computaiton errors
#print((sigma_f.transpose(0, 1) == sigma_f).all())

# make sigma_f symmetric
sigma_f_symmetric = (sigma_f + sigma_f.T)/2
#print((sigma_f_symmetric.transpose(0, 1) == sigma_f_symmetric).all())
#print(torch.all(sigma_f_symmetric.transpose(0, 1) == sigma_f_symmetric))

eigenvalues, eigenvectors = torch.linalg.eigh(sigma_f_symmetric)
```

We now have the blocks to build $\Sigma_f = Q\Lambda Q^T$ but we want $\Sigma_f = AA^T$. So we write $\Sigma_f = Q\Lambda^{1/2}\Lambda^{1/2}Q^T$ and thus we can define $A=Q\Lambda^{1/2}$. 

And ofc $Q$ is the $n\times n$ matrix whose $i$'th column corrosponds to the $i$'th eigenvector of $\Sigma_f$ and $\Lambda$ is a diagonal matrix whose entries corrosponds to the eigenvalues of $\Sigma_f$

Before we can build the matrix A, we want to make ensure positive semidefinit, since a valid covariance matrix must be postivie semidefinite. Thus we will creat the diagonal matrix $\Lambda$ with entry $i$ as $max(\lambda_i,\delta)$ with a very small, almost zero $\delta>0$, since some of the eigenvalues are below zero, which might be caused due to numerical precision errors.

```python
print(sum(eigenvalues<0))
```

```python
torch.manual_seed(123)

# Withdraw the eigenvalues and eigen vectors of our now symmetric covariance matrix 
eigenvalues, eigenvectors = torch.linalg.eigh(sigma_f_symmetric)

eigenvalues_non_negative = torch.where(eigenvalues > 0, eigenvalues, 1e-10)
A = eigenvectors @ torch.diag(eigenvalues_non_negative.sqrt())

# we can now draw the samples for the 3 different values of q_hat
eps_q1 = torch.distributions.Normal(0, 1).sample((5,101))
f_samples_q1 = mu_f_q1.T + eps_q1 @ A.T 

eps_q2 = torch.distributions.Normal(0, 1).sample((5,101))
f_samples_q2 = mu_f_q2.T + eps_q2 @ A.T 

eps_q3 = torch.distributions.Normal(0, 1).sample((5,101))
f_samples_q3 = mu_f_q3.T + eps_q3 @ A.T 

# let us plot the 5 samples from each of the three different q_hat's
plt.figure(figsize=(22, 6))

plt.subplot(1, 3, 1)
for i, sample in enumerate(f_samples_q1):
    plt.plot(x, sample, lw=3., label=f"sample_{i+1}")
plt.ylabel("f", fontsize=14)
plt.xlabel("x", fontsize=14)
plt.title("$\hat{q}$ = 0")
plt.legend()

plt.subplot(1, 3, 2)
for i, sample in enumerate(f_samples_q2):
    plt.plot(x, sample, lw=3., label=f"sample_{i+1}")
plt.ylabel("f", fontsize=14)
plt.xlabel("x", fontsize=14)
plt.title("$\hat{q}$ = 5")
plt.legend()

plt.subplot(1, 3, 3)
for i, sample in enumerate(f_samples_q3):
    plt.plot(x, sample, lw=3., label=f"sample_{i+1}")
plt.ylabel("f", fontsize=14)
plt.xlabel("x", fontsize=14)
plt.title("$\hat{q}$ = 10")
plt.legend()

plt.show()
```
